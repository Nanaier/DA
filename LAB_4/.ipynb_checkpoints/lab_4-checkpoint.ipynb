```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

1. Подивитись, проаналізувати структуру

Зчитуємо файл

```python
path = 'data/Data2.csv'

dataset = pd.read_csv(path, sep=';')
```

Очищаємо дані: заміна ком на крапки, заміна пустих значень на середні та заміна негативних значень на позитивні

```python
def convert_column_to_float(dataset, column_label):
    dataset[column_label] = dataset[column_label].str.replace(',', '.').astype(float)

convert_column_to_float(dataset, "Area")
convert_column_to_float(dataset, "GDP per capita")
convert_column_to_float(dataset, "CO2 emission")

def convert_nan_to_mean(dataset):
    dataset.fillna(dataset.mean(), inplace=True)

convert_nan_to_mean(dataset)

def convert_float_to_positive(dataset, column_label):
    dataset[column_label] = dataset[column_label].abs()

convert_float_to_positive(dataset, "GDP per capita")
convert_float_to_positive(dataset, "Area")
```

2. Вказати, чи є параметри, що розподілені за нормальним законом

Будуємо гістограми для візуальної оцінки. Бачимо, що не одна з гістограм параметрів не схожа за будовою на розподілену за нормальним законом.

```python
def create_hists(dataset):
    grp, axes = plt.subplots(2, 2, figsize=(12, 7))
    grp.suptitle("Histograms")
    axes[0, 0].hist(dataset["Area"])
    axes[0, 0].set_title("Area")

    axes[0, 1].hist(dataset["Population"])
    axes[0, 1].set_title("Population")

    axes[1, 0].hist(dataset["CO2 emission"])
    axes[1, 0].set_title("CO2 emission")

    axes[1, 1].hist(dataset["GDP per capita"])
    axes[1, 1].set_title("GDP per capita")
    plt.show()

create_hists(dataset)
```

Проводимо тести на нормальне розподілення за критеріями нормальності Пірсона та Шапіро

```python
def normality_test(column_label, test, a=0.05):
    statistics, p = test(column_label)
    print(f"Normality test result for '{column_label.name}': Statistics= '{statistics}', p= '{p}'")
    if p > a:
    	print('Data is normally distributed.\n')
    else:
    	print('Data is not normally distributed.\n')
```

```python
normality_test(dataset['Area'], stats.shapiro)
normality_test(dataset['Population'], stats.shapiro)
normality_test(dataset['CO2 emission'], stats.shapiro)
normality_test(dataset['GDP per capita'], stats.shapiro)
```

```python
normality_test(dataset['Area'], stats.normaltest)
normality_test(dataset['Population'], stats.normaltest)
normality_test(dataset['CO2 emission'], stats.normaltest)
normality_test(dataset['GDP per capita'], stats.normaltest)
```

3. Перевірити гіпотезу про рівність середнього і медіани для одного з параметрів

Перевіряємо гіпотезу про рівність середнього і медіани. Бачимо, що вони не співпадають ні для одного з параметрів

```python
def mean_median_significance(dataset, alpha = 0.05):
    N = len(dataset)
    df = N - 1
    deviation = dataset.std()
    median = dataset.median()
    mean = dataset.mean()

    t = abs(mean - median) / (deviation / math.sqrt(N))
    p_value = stats.t.sf(t, df)
    success = p_value > alpha / 2
    success_message = \
        f"{dataset.name}: Expected value and median are the same." if success else \
        f"{dataset.name}: Expected value and median are not the same."
    print(success_message)

mean_median_significance(dataset['Area'])
mean_median_significance(dataset['Population'])
mean_median_significance(dataset['CO2 emission'])
mean_median_significance(dataset['GDP per capita'])
```

4. Вказати, в якому регіоні розподіл викидів СО2 найбільш близький до нормального

Виведемо гістограми викидів по Регіонам

```python
dataset['CO2 emission'].hist(by=dataset['Region'], layout=(3, 3), figsize=(15, 10))

plt.show()
```

Перевіримо в якому з регіонів розподіл викидів СО2 найбільш близький до нормального

```python
regions = pd.unique(dataset['Region'])

for region in regions:
    region_emissions = dataset[dataset['Region'] == region]['CO2 emission']

    print(f"Checking for normal distribution of CO2 emissions in: {region}")
    try:
        normality_test(region_emissions, stats.shapiro)
        normality_test(region_emissions, stats.normaltest)
    except ValueError as e:
        print(str(e))
```

Судячи з результатів, бачимо, що не один з розподілів викидів по регіонам не є нормальним

5. Побудувати кругову діаграму населення по регіонам

Створюємо кругову діаграму населення по регіонам

```python
def show_pie_diagram(dataset, labels, column):
    title, diagrams = plt.subplots(figsize=(8, 6))
    diagrams.pie(dataset[column],
                 labels=labels, autopct='%1.1f%%')
    title.suptitle(f"Pie for {column}:", fontsize=20)

    plt.show()

show_pie_diagram(dataset.groupby("Region").sum(), regions, "Population")
```
